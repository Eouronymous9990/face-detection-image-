```python
import cv2
img=cv2.imread(r"E:\WhatsApp Image 2023-07-12 at 17.44.46.jpg")
import matplotlib.pyplot as plt
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
grey_frame=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur=cv2.GaussianBlur(grey_frame,(3,3),0)
face_detection=face_cascade.detectMultiScale(blur)
    
for x,y,w,h in face_detection:
    #
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
num_faces = len(face_detection)  # عدد الوجوه المكتشفة
text = f"Number of faces: {num_faces}"
cv2.putText(img, text, (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (165, 99, 255), 4)
    
#cv2.imshow('frame',img)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 8))
plt.imshow(img_rgb)
plt.title('Final Result')
plt.axis('off')
plt.show()
 
cv2.waitKey(0) 
cv2.destroyAllWindows()  
```
